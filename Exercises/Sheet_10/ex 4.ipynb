```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def GenerateData(N):
    X = np.repeat(np.array([[0,0],[0,1],[1,0],[1,1]]),N,axis=0)
    X = X + np.random.randn(4*N,2)*0.2
    y= np.repeat([1,-1,-1,1],N)
    return X,y
```

```python
X,y=GenerateData(100)
X=X-0.5
norm = np.linalg.norm(X) 
plt.scatter(X.T[0]*X.T[1],y)
```

```python
def phi(x_i):
    x_i=x_i-0.5
    return np.array(x_i[0]*x_i[1])
w=0.01
b=0.3
```

```python
def NN(x_i,w,b):
    return w*phi(x_i)+b
```

```python
def hingeLoss(x_i,y_i):
    print(np.maximum(0,1-NN(x_i,w,b)*y_i))
    return np.maximum(0,1-NN(x_i,w,b)*y_i)
```

```python
def gradient(x_i,y_i):
    if(hingeLoss(x_i,y_i)==0):
        return np.zeros(len(w)), 0
    else:
        return w/np.abs(w)phi(x_i)*y_i, b/np.abs(b) *y_i
```

```python
#Generate Data
X_train, Y_train = GenerateData(100)
X_test, Y_test = GenerateData(50)
for i in range(100):
    wg,bg  =gradient(X_train[i],Y_train[i])
    w=w- 0.5*wg
    b=b- 0.5*bg
   
```

```python
X_test,y_test=GenerateData(1)
print(X_test)
print(y_test)
```
