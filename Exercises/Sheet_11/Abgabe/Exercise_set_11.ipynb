In the following Exercise we will sample from a GP Prio with different kernels i.e.  Covariance matrices. We will implement a linear, a squared exponentialand a periodic Kernel. We will first load the necessary libraries:

```python
import numpy as np
import scipy.spatial
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.gridspec as gridspec
```

Next, we will define the three kernels:

```python
def linear_kernel(x,y):
    assert x.ndim==2, "everything should be arrays with ndim==2"
    assert y.ndim==2, "everything should be arrays with ndim==2"
    return x@y.T
# Define the exponentiated quadratic with σ=1 and bandwith l = 1
def exponentiated_quadratic(xa, xb,l=1):
    """Exponentiated quadratic  with σ=1"""
    # L2 distance (Squared Euclidian)
    σ=1
    sq_norm =  -1*scipy.spatial.distance.cdist(xa, xb, 'sqeuclidean')
    return (σ**2)*np.exp(sq_norm/(2*l))
#Define periodic kernel with with bandwith = sigma = 1 and p=2
def periodic(xa, xb):
    σ=1
    l=1
    p=4
    # L2 distance (absolut)
    abs_norm =  scipy.spatial.distance.cdist(xa, xb, 'cityblock')
    return (σ**2)*np.exp((-2/l)*np.sin(np.pi*(abs_norm/p)**2))
```

```python
# Sample from the Gaussian process distribution
nb_of_samples = 1000  # Number of points in each function
number_of_functions = 4  # Number of functions to sample
def SamplefromGP(kernel):
    # Independent variable samples
    X = np.expand_dims(np.linspace(0, 10, nb_of_samples), 1)
    Σ = kernel(X, X)  # Kernel of data points
    # Draw samples from the prior at our data points.
    # Assume a mean of 0 for simplicity
    ys = np.random.multivariate_normal(
    mean=np.zeros(nb_of_samples), cov=Σ, 
    size=number_of_functions)
    return X,ys

```

```python
def plot_GPs(X,y_s, title):
    # Plot the sampled functions
    plt.figure(figsize=(6, 4))
    for i in range(number_of_functions):
        plt.plot(X, y_s[i], linestyle='-', marker='o', markersize=3)
    plt.xlabel('$x$', fontsize=13)
    plt.ylabel('$y = f(x)$', fontsize=13)
    plt.title((
        '4 different function realizations at 41 points\n'
        'sampled from a Gaussian process with '+title))
    plt.xlim([0, 10])
    plt.show()
```

```python
#Sampling and Plotting with Linear Kernel
X,y_s = SamplefromGP(linear_kernel)
plot_GPs(X,y_s, "Linear Kernel")
#Sampling and Plotting with exponential Kernel
X,y_s = SamplefromGP(exponentiated_quadratic)
plot_GPs(X,y_s, "Squared-exponential Kernel")
#Sampling and Plotting with Periodic Kernel
X,y_s = SamplefromGP(periodic)
plot_GPs(X,y_s, "Periodic Kernel")
```

## Exercise 3 
We will first implement the alorithm 2.1 from the given article. We do not have to reimplement the squarred exponential covariance since we already implemented it above.

```python
X=np.array([0.5,1.1,1.7,2.3,2.9,3.5,4.1,4.7])[np.newaxis]
y=np.array([2.3,1.9,3.3,5.5,3.1,9.1,7.8,7.8])[np.newaxis]
X_t=np.array([3.4])[np.newaxis]
l=1.5
noise= 0.2
plt.scatter(X,y)
```

```python
def like_and_pred(X,y,k,noise,X_t):
    #L := cholesky(K + σ2nI)
    K= k(X.T,X.T,l)
    L=np.linalg.cholesky(K+np.identity(len(X.T))*noise**2)
    #α := L.T\(L\y)
    alpha=np.linalg.solve(L.T,np.linalg.solve(L,y.T))
    #f∗ := k.T∗ α
    mean_f=k(X.T,X_t,l).T@alpha
    #v := L\k∗
    v= np.linalg.solve(L,k(X.T,X_t,l))
    #V[f∗] := k(x∗, x∗) − v.Tv
    var_f= k(X_t,X_t,l)-v.T@v
    #og p(y|X) := − 12 y>α − ∑_i log Lii − n/2 log 2π 
    log_marg_like = -y@alpha/2-sum(np.log(np.diagonal(L)))-(len(X.T)/2)*np.log(2*np.pi)
    return mean_f,var_f,log_marg_like
```

We will now apply our model to the data givene for exercise b)

```python
mean_f,var_f,log_marg_like = like_and_pred(X,y,exponentiated_quadratic,noise,X_t)
print("Mean:",mean_f)
print("Variance", var_f)
print("Log-Marginal-Likelihood",log_marg_like)
```

We will now head to exercise c, where we will generate 100 testpoints and apply the GP to them. We will then plot the training data and there predicted mean.

```python
#Get Testpoints
X_test = np.sort(np.random.rand(100)*4.5 + 0.5)

#Initialize Arrays
means =[]
variances = []
LMLs = []
#Apply GP
for X_t in X_test :
    X_t = np.array([X_t])[np.newaxis]
    mean_f,var_f,log_marg_like = like_and_pred(X,y,exponentiated_quadratic,noise,X_t)
    means.append(mean_f[0][0])
    variances.append(var_f[0][0])
    LMLs.append(log_marg_like)
```

```python
#Calculate Intervalls for confidence intervalls
means_plus = means + 2*np.sqrt(variances)
means_minus = means - 2*np.sqrt(variances)

plt.plot(X_test, means, 'r-')
plt.fill_between(X_test, means_plus, means_minus, facecolor='gray')
plt.scatter(X,y)
plt.title(('Means and Training Data'))
plt.xlim([0, 5.5])
plt.savefig("EX3.png")
plt.show()
```
