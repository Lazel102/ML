```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, expon, gamma, laplace, norm
```

```python
x = norm.rvs(5, 1.5, size=1000)
no = norm.pdf(x, 5, 1.5)
plt.scatter(x, no)
plt.xlim([0, 10])
plt.ylim([0, 0.4])
plt.show()
```

```python
x = laplace.rvs(5, 1.5, 1000)
la = laplace.pdf(x, 5, 1.5)
plt.scatter(x, la)
plt.xlim([-5, 15])
plt.ylim([0, 2.5])
plt.show()
```

```python
x = expon.rvs(0, 1/1.5, 1000)
exp = expon.pdf(x, 1, 1/1.5)
plt.scatter(x, exp)
plt.xlim([-1, 8])
plt.ylim([-1, 3])
plt.show()
```

```python
x = gamma.rvs(5, 0, 1, 1000)
ga = gamma.pdf(x, 5, 0, 1)
plt.scatter(x, ga)
plt.xlim([0, 15])
plt.ylim([0, 0.3])
plt.show()
```

```python
x = beta.rvs(2, 1.5, 0, 1, 1000)
be = beta.pdf(x, 2, 1.5, 0, 1)
plt.scatter(x, be)
plt.xlim([0, 1])
plt.ylim([0, 2])
plt.show()
```
