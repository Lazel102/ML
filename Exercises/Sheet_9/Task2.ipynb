```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euclidean_distance(x, mean):
    return np.sqrt((x[0]- mean[0])**2 + (x[1]-mean[1])**2)
```

```python
def get_z(x, means):   
    distance = [euclidean_distance(x, mean) for mean in means]
    return np.argmin(distance)
```

```python
def sort_to_cluster(X, Z, k):
    cluster = 0
    sorted_clusters = []
    while(cluster != k):
        tmp_cluster = []
        for index, x in enumerate(X):
            if(Z[index]==cluster):
                tmp_cluster.append(x)
        cluster+=1;
        sorted_clusters.append(tmp_cluster)
    return sorted_clusters
```

```python
def calculate_new_mean(cluster):
    x_coords = []
    y_coords = []
    for coords in cluster:
        x_coords.append(coords[0])
        y_coords.append(coords[1])
    mean = [np.sum(x_coords)/len(cluster)]
    mean.append(np.sum(y_coords)/len(cluster))
    return mean
        
```

```python
def k_means(k, X):
    means = np.random.rand(k, 2) * 2
    z = [get_z(x, means) for x in X]
    tmp = 0
    while(z != tmp):
        tmp = z
        sorted_clusters = sort_to_cluster(X, z, k)
        means = [calculate_new_mean(cluster) for cluster in sorted_clusters]
        z = [get_z(x, means) for x in X]
    return means, sorted_clusters
    
```

```python
coordinates1 = np.random.rand(100, 2) * 2
coordinates2 = np.random.rand(100, 2) * 2 + 1.5
coords = np.concatenate((coordinates1, coordinates2), axis=0)
color = np.concatenate((np.zeros(100), np.ones(100)))
```

```python
plt.scatter(coords.T[0], coords.T[1], c=color)
plt.show()
```

```python
means, clusters = k_means(3, coords)
```

```python
plt.plot(means[0][0], means[0][1], marker='+')
plt.plot(means[1][0], means[1][1], marker='+')
plt.plot(means[2][0], means[2][1], marker='+')
plt.scatter(coords.T[0], coords.T[1], c=color)
plt.savefigure
plt.show()
```
